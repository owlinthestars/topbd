```python
plt.rcParams["figure.figsize"] = [10, 10]
plt.gca().invert_yaxis()

matrixm = np.random.randint(2, size=(95,95))
size = (95,95)
step = 5

for i in range(95):
    for j in range(95):
        x=[]
        y=[]
        x.append(j * step)
        y.append(i * step)
        
        if matrixm[i][j] == 0:
            x.append(x[0] + step)
            y.append(y[0])
        else:
            y.append(y[0] + step)
            x.append(x[0])
            
        plt.plot(x,y,'r')

plt.show()
```
